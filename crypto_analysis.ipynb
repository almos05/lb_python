### Работаем с pandas и matplotlib

```python
import math

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./datasets/crypto-markets.csv")[['date', 'open', 'low', 'high', 'close']][:200]
```

```python
df.head()
```

```python
mean = df['close'].mean()
std = df['close'].std()
median = df['close'].median()

plt.plot(df['close'], 'b', label='close')
plt.plot(df['low'], 'y--', label='low')
plt.plot(df['high'], 'g--', label='high')  # label="легенда линии"
plt.axhline(mean, color="red")
plt.axhline(mean + std, color='r', linestyle='--')
plt.axhline(mean - std, color='r', linestyle='--')
plt.axhline(median, color='blue', linestyle='-.')
plt.title('Рис. 1')  # название
plt.legend()  # рисуем легенду
plt.show()
```

### Numpy бом-бом-бом

```python
X = np.linspace(0, 100, 1000)
X[:5]
```

```python
plt.plot(X, np.sin(X), color='r')
plt.plot(X, np.cos(X), color='b')
plt.plot(X, np.cos(X) + np.sin(X), color='black')
plt.ylim(-2, 10)
plt.show()
```

### Оп, оп, Смешная аппроксимация

```python
data = df['close']
data.head()
```

```python
data_aprox = []

for i in range(len(data) - 1):
    data_aprox += [data[i], (data[i] + data[i + 1]) / 2]
data_aprox.append(data[len(data) - 1])

assert len(data_aprox) == 399

plt.plot(np.arange(0, 200 - 0.5, 0.5), data_aprox, color='black')
plt.plot(np.arange(200), data, color='r', linestyle='-.')
plt.show()
```
